## Question 1
For this question, calculate the win/loss ratio's correlation with the population of the city it is in for the **NHL** using **2018** data.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re



def nhl_correlation(): 
    
    nhl_df=pd.read_csv("assets/nhl.csv")
    cities=pd.read_html("assets/wikipedia_data.html")[1]
    cities=cities.iloc[:-1,[0,3,5,6,7,8]]
    
    cities['NHL']=cities['NHL'].str.replace(r'[*]','',regex=True)
    nhl_df['team']=nhl_df['team'].str.replace(r'[*]','')
    nhl_df=nhl_df[nhl_df['year']==2018]
    nhl_df['team']=nhl_df['team'].str.replace(r'[*]','')
    nhl_df=nhl_df.drop([0,9,18,26])
    nhl_df=nhl_df.reset_index()
    del nhl_df['index']

    cities['NHL']=cities['NHL'].str.replace(r'\[[^()]*\]','')
    cities['NFL']=cities['NFL'].str.replace(r'\[[^()]*\]','')
    cities['NBA']=cities['NBA'].str.replace(r'\[[^()]*\]','')
    cities['MLB']=cities['MLB'].str.replace(r'\[[^()]*\]','')
    cities=cities.rename(columns={'Population (2016 est.)[8]':'Population'})

    nhl=cities[['Metropolitan area','Population','NHL']]
    nhl['Population']=nhl['Population'].astype(int)
    nhl['NHL']=nhl['NHL'].replace('',np.nan)
    nhl['NHL']=nhl['NHL'].replace('—',np.nan)
    nhl=nhl.dropna().reset_index()
    del nhl['index']

    nhl_df=nhl_df[['team','W','L']]
    nhl_df['W'] = nhl_df['W'].astype(int)
    nhl_df['L'] = nhl_df['L'].astype(int)
    nhl_df['win/loss']=nhl_df['W']/(nhl_df['W']+nhl_df['L'])

    s=[]
    for i in nhl.iterrows():
        x=re.findall('([A-Z][a-z]*)',i[1].NHL)
        s.append(x)


    wll=[]

    for i in s:
        z=0
        wl=0
        for j in i:
            y=0
            for  k in nhl_df['team']:
                if j in k:
                    z=z+1
                    wl=nhl_df['win/loss'][y] +wl
                y=y+1
        wll.append(wl/z)

    nhl['win/loss']=wll
    

    
    # YOUR CODE HERE
    #raise NotImplementedError()
    
    population_by_region = nhl['Population'].tolist() # pass in metropolitan area population from cities
    win_loss_by_region = nhl['win/loss'].tolist() # pass in win/loss ratio from nhl_df in the same order as cities["Metropolitan area"]

    #assert len(population_by_region) == len(win_loss_by_region), "Q1: Your lists must be the same length"
    #assert len(population_by_region) == 28, "Q1: There should be 28 teams being analysed for NHL"
    
    return stats.pearsonr(population_by_region, win_loss_by_region)[0]
   

```
